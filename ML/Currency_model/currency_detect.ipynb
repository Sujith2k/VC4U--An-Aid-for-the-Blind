```
from keras.models import load_model
from google.colab.patches import cv2_imshow
from keras.preprocessing.image import ImageDataGenerator
import cv2
import numpy as np
model = load_model("gdrive/My Drive/cohort/currency.h5")
```

```
img =[] 
labels = ['10','20','50','100','200','500','2000']
ims = cv2.imread("gdrive/My Drive/cohort/Dataset/Currency_Train/20_10.jpg")
img.append(cv2.resize( ims , (150,150) ,interpolation = cv2.INTER_CUBIC ))

x = np.array(img)
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```
for batch in test_datagen.flow(x, batch_size =1):
    pred = [i for i in model.predict(batch)[0]]
    print(pred)
    break
print(labels[ pred.index(max(pred)) ])
cv2_imshow(ims)
```
